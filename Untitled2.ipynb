```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
df_train = pd.read_csv('C:/Users/shrut/OneDrive/Desktop/New folder/train.csv')
df_test = pd.read_csv('C:/Users/shrut/OneDrive/Desktop/New folder/test.csv')
df_gender_submission = pd.read_csv('C:/Users/shrut/OneDrive/Desktop/New folder/gender_submission.csv')

# Display the first few rows of the training dataset
print(df_train.head())

# Check for missing values
print(df_train.isnull().sum())

# Handle missing values
# Fill missing 'Age' with median age
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].median())

# Fill missing 'Embarked' with the most frequent port
df_train['Embarked'] = df_train['Embarked'].fillna(df_train['Embarked'].mode()[0])

# Drop 'Cabin' as it has too many missing values
df_train = df_train.drop(columns=['Cabin'])

# Convert 'Sex' to a numerical variable
df_train['Sex'] = df_train['Sex'].map({'male': 0, 'female': 1})

# Convert 'Embarked' to a numerical variable
df_train['Embarked'] = df_train['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

# Check for duplicates
print(df_train.duplicated().sum())

# Summary statistics
print(df_train.describe())

# Data types
print(df_train.dtypes)

# Histogram of Age
plt.figure(figsize=(10, 6))
plt.hist(df_train['Age'], bins=30, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Bar plot of the number of survivors
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df_train)
plt.title('Survivors Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Bar plot of survival by gender
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=df_train)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks([0, 1], ['Male', 'Female'])
plt.grid(True)
plt.show()


```
